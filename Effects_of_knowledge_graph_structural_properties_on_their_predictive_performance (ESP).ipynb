A continuación se ejecuta el ciclo de generación de todos los archivos de configuración. Cada configuración difiere de la anterior únicamente en el parámetro α o *s* de las relaciones que siguen la ditribución de Zipf. Todas las configuraciones fueron modeladas a partir del [archivo de de prueba](https://github.com/gbagan/gmark/blob/master/use-cases/test.xml) (`test.xml`) incluido por Gaban y sus colaboradores (2015) en el repositorio de gMark:

```python
%cd /content/KGResearch/configs

for i in range(config_num):

  # Parámetros específicos del archivo:

  file_name = "config" + str(i + 1) + ".xml";
  zipf_val = str((i + 1) * zipf_step);

  # Abrir archivo de configuración:

  config_file = open(file_name, 'w');

  # Especificación de cantidad de grafos generados:

  config_file.write("""<generator>
  <size>""" + str(graphs_by_config) + "</size>")

  # Especificaciones de tamaño para cada grafo:

  for j in range(graphs_by_config):
    config_file.write("""\n  <graph>
  \t<nodes>""" + str(graph_size) + """</nodes>
  </graph>""")
    
  # Predicados, tipos y esquemas de relaciones:

  config_file.write("""\n
  <predicates>
	  <size>4</size>
	  <alias symbol="0">authors</alias>
	  <proportion symbol="0">0.5</proportion>

	  <alias symbol="1">publishedIn</alias>
	  <proportion symbol="1">0.3</proportion>

	  <alias symbol="2">heldIn</alias>
	  <proportion symbol="2">0.01</proportion>

	  <alias symbol="3">extendedTo</alias>
	  <proportion symbol="3">0.19</proportion>
  </predicates>

  <types>
	  <size>5</size>
	  <alias type="0">researcher</alias>
	  <proportion type="0">0.5</proportion>

	  <alias type="1">paper</alias>
	  <proportion type="1">0.3</proportion>

	  <alias type="2">journal</alias>
	  <proportion type="2">0.1</proportion>

	  <alias type="3">conference</alias>
	  <proportion type="3">0.1</proportion>

	  <alias type="4">city</alias>
	  <fixed type="4">100</fixed>
  </types>

	<schema>

		<source type="0"> <!-- researcher -->
	  	<target type="1" symbol="0" edge_type="0"> <!-- authors a paper -->
	    	<indistribution type="gaussian">
	      	<mu>3</mu>
	      	<sigma>1</sigma>
	  		</indistribution>
				<outdistribution type="zipfian">
					<alpha>"""+ zipf_val +"""</alpha>
				</outdistribution>
			</target>
		</source>


		<source type="1"> <!-- paper -->
			<target type="3" symbol="1" edge_type="1"> <!-- published in a conference -->
				<indistribution type="gaussian">
					<mu>40</mu>
					<sigma>10</sigma>
				</indistribution>
				<outdistribution type="uniform">
					<min>1</min>
					<max>1</max>
				</outdistribution>
			</target>

			<target type="2" symbol="3" edge_type="2"> <!-- extended to a journal -->
				<indistribution type="zipfian">
					<alpha>"""+ zipf_val +"""</alpha>
				</indistribution>
				<outdistribution type="uniform">
					<min>0</min>
					<max>1</max>
				</outdistribution>
			</target>
		</source>


		<source type="3"> <!-- conference -->
			<target type="4" symbol="2" edge_type="3"> <!-- held in a city -->
				<indistribution type="zipfian">
					<alpha>"""+ zipf_val +"""</alpha>
				</indistribution>
				<outdistribution type="uniform">
					<min>1</min>
					<max>1</max>
				</outdistribution>
			</target>
		</source>
	</schema>


	<workload id="0" size="50">
		<!-- bounds for the size of these queries -->
		<size>
			<conjuncts min="3" max="4"/>	<!-- number of A^M where M is a multiplicity 1 or * -->
			<disjuncts min="1" max="3"/>	<!-- number of w_1 + ... + w_n in each A -->
			<length min="2" max="4"/>		<!-- number of a_1 . ... . a_m in each w_i -->
		</size>

		<!-- probability to have "*" as multiplicity of the conjuncts -->
		<multiplicity star="0.5"/>

		<!-- the arity of the generated queries, we allow 0 for Boolean queries -->
		<arity min="0" max="4"/>

		<!-- we want only linear queries, but not constant nor quadratic -->
		<selectivity constant="1" linear="1" quadratic="1"/>

		<!-- we want to have in the workload chains and stars, but not cycles, etc. -->
		<type chain="1" star="1" cycle="1" starchain="1"/>
	</workload>""");

  config_file.write("\n\n</generator>");
  config_file.close();
```

### Aplicación y evaluación de TransE sobre grafos generados en (1.2)

Ciclo de aplicación de **TransE**, un modelo básico de aprendizaje automatizado y producción de embebidos sobre grafos de conocimiento. Por cada iteración del ciclo de aplicación, se considera uno de los grafos sintetizados de una de las configuraciones iniciales generados en las secciones [1.1](https://colab.research.google.com/drive/19ExEC9jP56yx5Snj8qM5PJpw2Sn_JMX1#scrollTo=A0Aog7ELtwAt) y [1.2](https://colab.research.google.com/drive/19ExEC9jP56yx5Snj8qM5PJpw2Sn_JMX1#scrollTo=piGKkSxa1Pk2), respectivamente. A partir de los subgrafos de entrenamiento, evaluación y validación seleccionados [anteriormente](https://colab.research.google.com/drive/19ExEC9jP56yx5Snj8qM5PJpw2Sn_JMX1#scrollTo=cB5Th8_6w_6X&line=3&uniqifier=1), se genera y evalúa una instancia de TransE para el grafo considerado. Como parte de dicha evaluación, se producen las sigueintes métricas de rendimiento:

*   **Media del rango** (MR): Media de la posición que alcanzó un triple verdadero, al ser comparado en probabilidad de veracidad con respecto a un conjunto de triples negativos sintéticos. A menor rango, mejor rendimiento del modelo.



*   **Media del rango recíproco** (MRR): Media del recíproco de la posición que alcanzó un triple verdadero, al compararse con negativos sintéticos. A mayor rango recíproco, mejor rendimiento del modelo.



*   **Hits@n**: Proporción de triples verdaderos que alcanzaron la posición *n* o una posición mejor al ser comparados con negatios sintéticos. A mayor proporción, mejor rendimiento del modelo. En este experimento, se consideran las **posiciones 1, 10 y 100**.

Al final de cada iteración del ciclo de configuraciones, se produce una tabla con las métricas resultantes de los modelos obtenidos para dicha configuración. Tras la compleción del ciclo de aplicación total, se produce adicionalmente otra tabla, con las métricas obtenidas para todas las instancias generadas de TransE. Estas tablas se guardan en el subdirectorio **KGResearch/metrics**.



```python
# Entrar a subdirectorio de aplicaciones de TransE:

%cd /content/KGResearch/models/TransE

# Importar bibliotecas relevantes para la aplicación del modelo:

import pandas as pd
from ampligraph.latent_features import TransE
from ampligraph.utils import save_model, restore_model
from ampligraph.evaluation import mr_score, mrr_score, hits_at_n_score

# Listas con instancias, rangos y métricas obtenidos para TransE:

graphs_TransE = [];
graphs_TransE_ranks = [];
graphs_TransE_mrs = [];
graphs_TransE_mrrs = [];
graphs_TransE_hits_1 = [];
graphs_TransE_hits_10 = [];
graphs_TransE_hits_100 = [];

# Listas con identificadores para grafos y configuraciones:

id_graphs = list(range(0, graphs_by_config));
id_graphs2 = [];
id_configs = [];

# Por cada configuración construida en (1.1),

counter = 0;
for i in range(len(configs)):

  # Listas con métricas de configuración:

  config_mrs = [];
  config_mrrs = [];
  config_hits_1 = [];
  config_hits_10 = [];
  config_hits_100 = [];

  # Por cada grafo sintetizado a partir de dicha configuración en (1.2),

  for j in range(graphs_by_config):

    print("\n---------------------------------------------------------")
    print("\n" + str(counter + 1) + ". Aplicación de TransE a grafo " 
               + str(j) + " de config" + str(i + 1) + ":\n");

    # Obtener nombres de los archivos originales del grafo:

    graph_txt = graphs_txts[counter];
    graph_csv = graphs_csvs[counter];
  
    # Leer datos del CSV y dividirlos de acuerdo a la forma sujeto-predicado-objeto:

    dataset = pd.read_csv(graph_txt, header=None)
    dataset.to_csv(graph_csv, index=None)
    dataset = dataset[0].str.split(' ', expand=True)
  
    dataset.columns = ['subject', 'predicate', 'object']
    dataset.head(5)

    # Obtener conjuntos de triples de entrenamiento, validación y evaluación:

    X_train = graphs_train[counter];
    X_valid = graphs_valid[counter];
    X_test = graphs_test[counter];

    # Imprimir dimensiones de cada selección:
  
    print('Total triples:', dataset.shape)
    print('Size of train:', X_train.shape)
    print('Size of valid:', X_valid.shape)
    print('Size of test:', X_test.shape)

    # Entrenar modelo TransE sobre los datos:

    model = TransE(k=150,                                                           # embedding size
                 epochs=100,                                                        # Num of epochs
                 batches_count= 10,                                                 # Number of batches 
                 eta=1,                                                             # number of corruptions to generate during training
                 loss='pairwise', loss_params={'margin': 1},                        # loss type and it's hyperparameters         
                 initializer='xavier', initializer_params={'uniform': False},       # initializer type and it's hyperparameters
                 regularizer='LP', regularizer_params= {'lambda': 0.001, 'p': 3},   # regularizer along with its hyperparameters
                 optimizer= 'adam', optimizer_params= {'lr': 0.001},                # optimizer to use along with its hyperparameters
                 seed= j, verbose=True)
  
    model.fit(X_train)

    # Guardar modelo:

    model_file = 'TransE-config' + str(i + 1) +'-graph' + str(j) + '.pkl'
    save_model(model, model_file)
    graphs_TransE.append(model_file)
    print("\nSaved model of config" + str(i + 1) + " graph to '" + model_file + "'\n")

    # Adquirir filtro para evaluar el modelo:
  
    X_filter = graphs_filters[counter];
  
    # Computar rangos y desplegar métricas:
  
    ranks = evaluate_performance(X_test, 
                                 model=model, 
                                 filter_triples=X_filter)
    graphs_TransE_ranks.append(ranks)
  
    print("Test set for model 'model_file':", X_test.shape)
    print("Size of ranks:", ranks.shape)
  
    display_aggregate_metrics(ranks)

    # Guardar métricas e identificadores en listas de configuración:

    config_mrs.append(mr_score(ranks));
    config_mrrs.append(mrr_score(ranks));
    config_hits_1.append(hits_at_n_score(ranks, n=1));
    config_hits_10.append(hits_at_n_score(ranks, n=10));
    config_hits_100.append(hits_at_n_score(ranks, n=100));

    counter += 1;
    id_configs.append(i + 1);
    id_graphs2.append(j);

  # Construir y guardar tabla con métricas de configuración:

  Config_metrics = pd.DataFrame(list(zip(id_graphs,
                                         config_mrs,
                                         config_mrrs, 
                                         config_hits_1,
                                         config_hits_10,
                                         config_hits_100)),
                                columns = ['Graph',
                                         'Mean Rank', 'Mean Reciprocal Rank', 
                                         'Hits@1', 'Hits@10', 'Hits@100'])
    
  Config_metrics.to_csv('/content/KGResearch/metrics/TransE_config' + str(i + 1) + '.csv');

  # Actualizar listas de métricas obtenidas al aplicar TransE:

  graphs_TransE_mrs = graphs_TransE_mrs + config_mrs;
  graphs_TransE_mrrs = graphs_TransE_mrrs + config_mrrs;
  graphs_TransE_hits_1 = graphs_TransE_hits_1 + config_hits_1;
  graphs_TransE_hits_10 = graphs_TransE_hits_10 + config_hits_10;
  graphs_TransE_hits_100 = graphs_TransE_hits_100 + config_hits_100;

# Construir y guardar tabla con métricas de aplicación de TransE:

TransE_metrics = pd.DataFrame(list(zip(id_configs, 
                                       id_graphs2,
                                       graphs_TransE_mrs,
                                       graphs_TransE_mrrs, 
                                       graphs_TransE_hits_1,
                                       graphs_TransE_hits_10,
                                       graphs_TransE_hits_100)),
                                columns = ['Conf.', 'Graph',
                                         'Mean Rank', 'Mean Reciprocal Rank', 
                                         'Hits@1', 'Hits@10', 'Hits@100'])
    
TransE_metrics.to_csv('/content/KGResearch/metrics/TransE_metrics.csv');
```

Ciclo de aplicación de **ComplEx**, un modelo de aprendizaje automatizado y producción de embebidos sobre grafos de conocimiento que emplea multiplicación vectorial sobre el cuerpo complejo. Por cada iteración del ciclo de aplicación, se considera uno de los grafos sintetizados de una de las configuraciones iniciales generados en las secciones [1.1](https://colab.research.google.com/drive/19ExEC9jP56yx5Snj8qM5PJpw2Sn_JMX1#scrollTo=A0Aog7ELtwAt) y [1.2](https://colab.research.google.com/drive/19ExEC9jP56yx5Snj8qM5PJpw2Sn_JMX1#scrollTo=piGKkSxa1Pk2), respectivamente. A partir de los subgrafos de entrenamiento, evaluación y validación seleccionados [anteriormente](https://colab.research.google.com/drive/19ExEC9jP56yx5Snj8qM5PJpw2Sn_JMX1#scrollTo=cB5Th8_6w_6X&line=3&uniqifier=1), se genera y evalúa una instancia de ComplEx para el grafo considerado. Al final de cada iteración del ciclo de configuraciones, se produce una tabla con las métricas resultantes (media del rango, media del rango recíproco, Hits@1, Hits@10 y Hits@100) de los modelos obtenidos para dicha configuración. Tras la compleción del ciclo de aplicación total, se produce otra tabla con las métricas obtenidas para todas las instancias generadas de ComplEx. Estas tablas se guardan en el subdirectorio **KGResearch/metrics**.

```python
# Entrar a subdirectorio de aplicaciones de ComplEx:

%cd /content/KGResearch/models/ComplEx

# Importar bibliotecas relevantes para la aplicación del modelo:

from ampligraph.latent_features import ComplEx

# Listas con instancias, rangos y métricas obtenidos para ComplEx:

graphs_ComplEx = [];
graphs_ComplEx_ranks = [];
graphs_ComplEx_mrs = [];
graphs_ComplEx_mrrs = [];
graphs_ComplEx_hits_1 = [];
graphs_ComplEx_hits_10 = [];
graphs_ComplEx_hits_100 = [];

# Listas con identificadores para grafos y configuraciones:

id_graphs = list(range(0, graphs_by_config));
id_graphs2 = [];
id_configs = [];

# Por cada configuración construida en (1.1),

counter = 0;
for i in range(len(configs)):

  # Listas con métricas de configuración:

  config_mrs = [];
  config_mrrs = [];
  config_hits_1 = [];
  config_hits_10 = [];
  config_hits_100 = [];

  # Por cada grafo sintetizado a partir de dicha configuración en (1.2),

  for j in range(graphs_by_config):

    print("\n---------------------------------------------------------")
    print("\n" + str(counter + 1) + ". Aplicación de ComplEx a grafo " 
               + str(j) + " de config" + str(i + 1) + ":\n");

    # Obtener nombres de los archivos originales del grafo:

    graph_txt = graphs_txts[counter];
    graph_csv = graphs_csvs[counter];
    
    # Leer datos del CSV y dividirlos de acuerdo a la forma sujeto-predicado-objeto: 

    dataset = pd.read_csv(graph_txt, header=None)
    dataset.to_csv(graph_csv, index=None)
    dataset = dataset[0].str.split(' ', expand=True)
    dataset.columns = ['subject', 'predicate', 'object']

    # Obtener conjuntos de triples de entrenamiento, validación y evaluación:

    X_train = graphs_train[counter];
    X_valid = graphs_valid[counter];
    X_test = graphs_test[counter];

    # Imprimir dimensiones de cada selección:
    
    print('Total triples:', dataset.shape)
    print('Size of train:', X_train.shape)
    print('Size of valid:', X_valid.shape)
    print('Size of test:', X_test.shape, '\n')

    # Entrenar modelo ComplEx sobre los datos:

    model = ComplEx(k=150, epochs=50, eta=1, loss='multiclass_nll', 
                    initializer='xavier', initializer_params={'uniform': False},
                    regularizer='LP', regularizer_params= {'lambda': 0.0001, 'p': 3},
                    optimizer= 'adam', optimizer_params= {'lr': 0.001}, 
                    seed= j, batches_count= 1, verbose=True)
    
    model.fit(X_train)

    # Guardar modelo:

    model_file = 'ComplEx-config' + str(i + 1) +'-graph' + str(j) + '.pkl'
    save_model(model, model_file)
    graphs_ComplEx.append(model_file)
    print("\nSaved model of config" + str(i + 1) + "graph to '" + model_file + "'\n")

    # Adquirir filtro para evaluar el modelo:
    
    X_filter = graphs_filters[counter];
    
    # Computar rangos y desplegar métricas:
    
    ranks = evaluate_performance(X_test, 
                                model=model,
                                filter_triples=X_filter,
                                corrupt_side='s,o',
                                ranking_strategy='worst')
    graphs_ComplEx_ranks.append(ranks)
  
    print("Test set for model 'model_file':", X_test.shape)
    print("Size of ranks:", ranks.shape)
  
    display_aggregate_metrics(ranks)

    # Guardar métricas e identificadores en listas de configuración:

    config_mrs.append(mr_score(ranks));
    config_mrrs.append(mrr_score(ranks));
    config_hits_1.append(hits_at_n_score(ranks, n=1));
    config_hits_10.append(hits_at_n_score(ranks, n=10));
    config_hits_100.append(hits_at_n_score(ranks, n=100));

    counter += 1;
    id_configs.append(i + 1);
    id_graphs2.append(j);

  # Construir y guardar tabla con métricas de configuración:

  Config_metrics = pd.DataFrame(list(zip(id_graphs,
                                         config_mrs,
                                         config_mrrs, 
                                         config_hits_1,
                                         config_hits_10,
                                         config_hits_100)),
                                columns = ['Graph',
                                         'Mean Rank', 'Mean Reciprocal Rank', 
                                         'Hits@1', 'Hits@10', 'Hits@100'])
    
  Config_metrics.to_csv('/content/KGResearch/metrics/ComplEx_config' + str(i + 1) + '.csv');

  # Actualizar listas de métricas obtenidas al aplicar ComplEx:

  graphs_ComplEx_mrs = graphs_ComplEx_mrs + config_mrs;
  graphs_ComplEx_mrrs = graphs_ComplEx_mrrs + config_mrrs;
  graphs_ComplEx_hits_1 = graphs_ComplEx_hits_1 + config_hits_1;
  graphs_ComplEx_hits_10 = graphs_ComplEx_hits_10 + config_hits_10;
  graphs_ComplEx_hits_100 = graphs_ComplEx_hits_100 + config_hits_100;

# Construir y guardar tabla con métricas de aplicación de ComplEx:

ComplEx_metrics = pd.DataFrame(list(zip(id_configs, id_graphs2,
                                         graphs_ComplEx_mrs,
                                         graphs_ComplEx_mrrs, 
                                         graphs_ComplEx_hits_1,
                                         graphs_ComplEx_hits_10,
                                         graphs_ComplEx_hits_100)),
                                columns = ['Conf.', 'Graph',
                                         'Mean Rank', 'Mean Reciprocal Rank', 
                                         'Hits@1', 'Hits@10', 'Hits@100'])
    
ComplEx_metrics.to_csv('/content/KGResearch/metrics/ComplEx_metrics.csv');
```
